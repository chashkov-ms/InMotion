```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

from Triangulate.Triangulate import TDoA
```

# Определение используемых функций

```python
def plot_grid():
    plt.minorticks_on()
    #  Определяем внешний вид линий основной сетки:
    plt.grid(which='major',
        color = 'k', 
        linewidth = 1)

    #  Определяем внешний вид линий вспомогательной сетки:
    plt.grid(which='minor', 
        color = 'k', 
        linestyle = ':')
```

```python
def get_distance_tdoa(distance):
    row = []
    distance_matrix = []
    for i in range(len(distance)):
        for j in range(len(distance)):
            row.append(distance[i]-distance[j])
        distance_matrix.append(row.copy())
        row.clear()
    return np.array(distance_matrix)
```

```python
def get_distance_tdoa_noise(distance, sko = 0.2):
    
    noise = np.random.normal(0,sko,len(distance))
    distance = distance + noise
    row = []
    distance_matrix = []
    for i in range(len(distance)):
        for j in range(len(distance)):
                row.append(distance[i]-distance[j])
        distance_matrix.append(row.copy())
        row.clear()
    return np.array(distance_matrix)
```

```python
def get_measure(anker_coord, cord_x, coord_y):
    measure = []
    coord = np.array([cord_x, coord_y])
    coord = (anker_coord - coord)**2
    for anker in coord:
        measure.append(np.sqrt(anker.sum()))
    return np.array(measure)
```

```python
def test_data_spyral():
    r = [0]
    fi = [0]
    d_fi = 11/180*math.pi
    a = 25
    for i in range(110):
        fi.append(fi[-1]+d_fi)
        r.append(a/(2*math.pi)*fi[-1])
    r = np.array(r)
    fi = np.array(fi)

    x = np.cos(fi)*r
    y = np.sin(fi)*r
    return [x+70, y+60]
```

```python
def test_data_rand(n=1):
    x = np.random.sample(n)*150
    y = np.random.sample(n)*100
    return [x, y]
```

```python
def find_SKO(data, tdoa_data):

    sygma_x = (data[0]-tdoa_data[0])**2
    sygma_x = sygma_x[~np.isnan(sygma_x)]
    sygma_x = sygma_x.sum()/len(sygma_x)
    sygma_x = np.sqrt(sygma_x)
    sygma_y = (data[1]-tdoa_data[1])**2
    sygma_y = sygma_y[~np.isnan(sygma_y)]
    sygma_y = sygma_y.sum()/len(sygma_y)
    sygma_y = np.sqrt(sygma_y)
    return [sygma_x, sygma_y]
```

# Проверка отдельных точек (отладка алгоритма)

```python
anker_coord = np.array([[0, 0], [0, 100], [150, 100], [150, 0]])
measure = np.array([58, 86, 121, np.nan])
measure = get_measure(anker_coord, 50, 30)

distance = get_distance_tdoa(measure)

tdoa_tr = TDoA(anker_coord)
tdoa_tr.triangulate(distance)
```

```python
measure = get_measure(anker_coord, 90, 90)

distance = get_distance_tdoa(measure)

tdoa_tr.clear()
tdoa_tr.triangulate(distance)
```

# Тест спираль

## Спираль без шумов

```python
data = test_data_spyral()
tdoa_tr.clear()
tdoa_x = []
tdoa_y = []

for i in range(len(data[0])):
    measure = get_measure(anker_coord, data[0][i], data[1][i])
    distance = get_distance_tdoa(measure)
    coord = tdoa_tr.triangulate(distance)
    tdoa_x.append(coord[0])
    tdoa_y.append(coord[1])

SKO = find_SKO(data, [np.array(tdoa_x), np.array(tdoa_y)])
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.subplot(1, 2, 1)
plt.scatter(anker_coord[:,0], anker_coord[:,1], 
            label='anker', 
            marker = 'x', 
            color = 'blue')
plt.plot(data[0], data[1], label='data', marker = '.', color = 'red')

plt.xlabel('X coord, m')
plt.ylabel('Y coord, m')
plt.title ('Spyral')
           
plt.legend(loc='best')    
plot_grid()

plt.subplot(1, 2, 2)
plt.scatter(anker_coord[:,0], anker_coord[:,1], 
            label='anker', 
            marker = 'x', 
            color = 'blue')
plt.plot(tdoa_x, tdoa_y, label='data', marker = '.', color = 'red')

plt.xlabel('X coord, m')
plt.ylabel('Y coord, m')
plt.title ('Spyral_TDoA \nSKO_X = {}, \nSKO_Y = {}'.format(SKO[0], SKO[1]))
           
plt.legend(loc='best')    
plot_grid()
```

## Спираль с шумами измерения

```python
data = test_data_spyral()
tdoa_tr.clear()
tdoa_x_n = []
tdoa_y_n = []

for i in range(len(data[0])):
    measure = get_measure(anker_coord, data[0][i], data[1][i])
    distance = get_distance_tdoa_noise(measure)
    coord = tdoa_tr.triangulate(distance)
    tdoa_x_n.append(coord[0])
    tdoa_y_n.append(coord[1])
    
SKO = find_SKO(data, [np.array(tdoa_x_n), np.array(tdoa_y_n)])
```

# Тест одна случайная точка

```python
data = [np.random.randint(0, 150, 1), np.random.randint(0, 100, 1)]

tdoa_x_n = []
tdoa_y_n = []
use_data_x = []
use_data_y = []

for i in range(1000):
    tdoa_tr.clear()
    measure = get_measure(anker_coord, data[0][0], data[1][0])
    distance = get_distance_tdoa_noise(measure)
    coord = tdoa_tr.triangulate(distance)
    tdoa_x_n.append(coord[0])
    tdoa_y_n.append(coord[1])
    use_data_x.append(data[0])
    use_data_y.append(data[1])
    
SKO = find_SKO([np.array(use_data_x), np.array(use_data_y)], 
               [np.array(tdoa_x_n), np.array(tdoa_y_n)]
              )
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.scatter(anker_coord[:,0], anker_coord[:,1], 
            label='anker', 
            marker = 'x', 
            color = 'blue')
plt.scatter(tdoa_x_n, 
            tdoa_y_n, 
            label='tdoa_noise', 
            marker = '.', 
            color = 'black')
plt.scatter(data[0], 
            data[1], 
            label='data', 
            marker = '.', 
            color = 'red')

plt.xlabel('X coord, m')
plt.ylabel('Y coord, m')
plt.title ('One random point\nSKO_X = {}, \nSKO_Y = {}'.format(SKO[0], 
                                                               SKO[1])
          )
           
plt.legend(loc='best')    
plot_grid()
```
